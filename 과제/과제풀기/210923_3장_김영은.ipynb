<font color='green'> Load the necessary libraries </font>

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
import torch.nn.init as init
import matplotlib.pyplot as plt
```

<font color='green'> Data making </font>

```python
# 갯수는 임의 지정
num_data = 1000
```

```python
# x는 -10에서 10까지의 숫자들이 무작위로 들어가게 균등하게 초기화
x = init.uniform_(torch.Tensor(num_data,1),-10,10)
x.shape
```

```python
# 연산 그래프 정의 (연산을 통해 값이 초기화)
# y=2x+3

y = 2*x + 3
```

```python
# 노이즈 지정(mean = 디폴트 0, std = 0.2, init_normal이용하며 y와 같은 모양을 가져야함)
noise = init.normal_(torch.Tensor(num_data,1),std=0.2,mean=0)
noise.shape
```

```python
# y에 노이즈를 더해주기
# 실제 데이터에서 노이즈가 추가되는 경우가 많음
y_noise = y + noise
```

<font color ='green'>Data visualization</font>

```python
#데이터 시각화해보기
plt.figure(figsize = (10,10))
plt.scatter(x.numpy(),y.numpy(),s = 7, c = 'gray')
plt.axis([-12, 12, -25, 25])
plt.show()
```

<font color ='green'>Model</font>

```python
# 입력 x = (1000, 1)
# y= (1000, 1) 즉, 특성의 갯수는 하나 Linear(x,y)
model = nn.Linear(1,1)

# 손실함수 지정
# 임의로 L1사용
loss_func = nn.L1Loss()

# 최적화 함수 지정
# 인수로 학습의 대상이 되는 모델의 변수 전달
# 학습률 임의로 지정(0.01)
optimizer = optim.SGD(model.parameters(),lr=0.01)
```

Q. L1 loss와 MSE loss의 정의는?

A. 

L1 loss = 예측값과 실제값 차이의 절대값의 평균

MSE loss = 예측값과 실제값 차이의 제곱의 평균

<font color ='green'>Model Training</font>

```python
# loss가 어떻게 변하는지 확인하기 위해 loss_arr을 만들어 기록
loss_arr = []

# target은 y
label = y

# 500번 학습
num_epoch = 1000

for i in range(num_epoch):
    
    # 이전 기울기 지우고 최적화 함수 초기화
    # 기울기를 지우지 않으면 학습이 잘 되지 않음(그때그때 구해야 함)
    optimizer.zero_grad()
    
    # 입력값 trian['x']를 모델에 넣어 결과값을 얻음
    # nn.Linear(1, 1)(x)
    output = model(x)
    
    # 결과값과 목표값의 차이를 L1 손실 함수로 구해줌
    loss = loss_func(output,label)
    
    # 손실에 대한 기울기를 구함
    loss.backward()
    
    # 구한 기울기를 이용해 모델의 변수를 업데이트
    optimizer.step()
    
    
    # 10번마다 모델의 변수가 어떻게 변하고 있는지 시각화
    if i % 10 == 0:
        
        plt.scatter(x.detach().numpy(), output.detach().numpy())
        plt.axis([-10, 10, -30, 30])
        plt.show()
        params = list(model.parameters())
        W = params[0].item()
        b = params[1].item()
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
            i, num_epoch, W, b, loss))
    loss_arr.append(loss.detach().numpy())
```

```python
plt.figure(figsize=(15,15))
plt.scatter(x.numpy(),y.numpy(),c="blue")
plt.plot(x.detach().numpy(),output.detach().numpy(),c="red")
plt.axis([-10, 10, -30, 30])
plt.show()
```

<font color ='green'>Loss graph </font>

```python
plt.plot(loss_arr)
plt.show()
```
